#### 1. 2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = 0
df.loc[df['yr_renovated'] != 0,'delta_renovated'] = df['yr_renovated'] - df['yr_built']
df[df['yr_renovated'] != 0].sample(6) 
```

#### 1. 3 Создайте признаки года продажи, месяца продажи

```python
##на случай ошибки, тут drop
# df.drop(columns='sale_year',inplace=True)
# df.drop(columns='sale_month',inplace=True)

df.insert(2,'sale_year',df.date.str[:4])
df.insert(3,'sale_month',df.date.str[4:6])
df
```

#### 1. 4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date','zipcode','lat','long'],inplace=True)
df
```

```python
#Попробуем переместить нашу delta_renovated c конца поближе к соответствующим колонкам
#нагуглил метод df.columns.tolist() , идея такая , чтобы не писать вручную все названия, а работать со срезами по list(columns)

col_slice = df.columns.to_list()
place_to_insert = df.columns.get_loc("yr_renovated")+1
new_order = col_slice[0:place_to_insert]+col_slice[-1:] + col_slice[place_to_insert:-1]
# print(col_slice)
# new_order
df = df[new_order]
df
```

#### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
import numpy as np
#для более наглядного отображение(вместо научного), можно через astype указать другой тип отдельной колонке
df.pivot_table(index='bedrooms',
                values='price',
                aggfunc=np.mean,
                margins=True,
                margins_name='any'
                ).sort_values(by='price').astype({'price':int})

#другой вариант на пробу 
# with pd.option_context('display.float_format', lambda x: f'{x:,.3f}'):
    # display(your_df)
```

#### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
tab3 = pd.crosstab(index=df['condition'],
            columns=df['grade'],
            values=df['price'],
            aggfunc=['median'],
            margins_name='Общая',
            margins= True).fillna(0).astype(int)
tab3
```
