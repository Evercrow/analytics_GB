### Задача 1
*Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.*

Используем формулы для нахождения мат.ожидания и дисперсии для равномерного распределения:
$$
M(A)=\frac{a+b}{2}=\frac{200+800}{2}=500
$$
Математическое ожидание и будет равно среднему значению величины на данном промежутке.  
Теперь формула дисперсии:
$$

D(A)=\frac{(b-a)^2}{12}=\frac{(200-800)^2}{12}=30000
$$

### Задача 2
*О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.*

Да, можем, так как получаем уравнение с одной неизвестной, подставив все условия в формулу дисперсии равномерной величины.  
$$D(B)=\frac{(b-a)^2}{12}$$
$$0.2 = \frac{(b-0.5)^2}{12}$$
$$b-0.5= \sqrt{12*0.2}$$
$$b= 1.549+0.5 = 2.049 \approx 2.05$$

Среднее значение величины B будет равна мат.ожиданию по формуле:
$$M(B)=\frac{a+b}{2}=\frac{0.5+2.05}{2}=1.275$$

### Задача 4
*Рост взрослого населения города X имеет нормальное распределение.  
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.  
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:    
а). больше 182 см  
б). больше 190 см  
в). от 166 см до 190 см  
г). от 166 см до 182 см  
д). от 158 см до 190 см  
е). не выше 150 см или не ниже 190 см  
ё). не выше 150 см или не ниже 198 см  
ж). ниже 166 см.*

используем формулу нормирования (оценка отклонения в сигмах) и таблицу накопленного нормального распределения:
$ Z = \frac{X-\mu}{\sigma} $ ,где по условию задачи $\sigma = 8$ см, и  $\mu=174$ см


**а). больше 182 см**  
$ Z = \frac{182-174}{8}= 1 $.  
Смотрим по таблице значение  по [1.0] и [0.00], $P(x \leq 182)= 0.8413$  
Значит,  $P(x \geq 182)=1- 0.8413= 0.1587$  
Или же можем пойти по правилу 3-х сигм - отклонение вправо на одну сигма, значит, процент значений в пределах от 182 см будет:  
$100\%- (50\% $(слева от $\mu$) $+ \frac{68\%}{2}) = 100\%-84\%=16\%$ . 

**б). больше 190 см** 
 $ Z = \frac{190-174}{8}= 2 $.  
Смотрим по таблице значение  по [2.0] и [0.00], $P(x \leq 182)= 0.9772$  
Значит,  $P(x \geq 190)=1- 0.9772= 0.0228$  
По правилу 3-х сигм - отклонение вправо на два сигма, значит, процент значений в пределах от 190 см будет:  
$100\%- (50\% $(слева от $\mu$) $+ \frac{95.4\%}{2}) = 100\%-97.7\%=2.3\%$ . 

**в). от 166 см до 190 см**  
$ Z = \frac{166-174}{8}= -1 $.  
Смотрим по таблице значение  по [-1.0] и [0.00], $P(x \leq 166)= 0.1587$  
Значит,  $P(x \geq 166)=1- 0.1587= 0.8413$  
Отсюда вычтем значения после 190 см, это мы посчитали в **(б)**  
$P(166см \leq x \leq 190 см)= 0.8413 - 0.0228 = 0.8185$  
По правилу 3-х сигм - отклонение влево на 1 сигма и вправо на 2, значит, процент значений в пределах $166см \leq x \leq 190 см$ равен:  
$\frac{68\%}{2} + \frac{95.4\%}{2} = 34\%+47.7\%=81.7\%$ . 

**г). от 166 см до 182 см**  
Из оценок ранее мы уже знаем,что от 166см до 182см - это отрезок $[-\sigma,+\sigma]$
Воспользуемся правилом 3-х сигм, и сразу дадим ответ: $68\%$ всех значений


**д). от 158 см до 190 см**  
$ Z = \frac{158-174}{8}= -2 $.  
Смотрим по таблице значение  по [-2.0] и [0.00], $P(x \leq 158)= 0.0228$  
Значит,  $P(x \geq 166)=1- 0.0228 = 0.9772$  
Отсюда вычтем значения после 190 см,это на графике такой же огрызок по сигма, как и слева  
$P(158 см \leq x \leq 190 см)= 0.9772 - 0.0228 = 0.9544$  
По правилу 3-х сигм - отклонение влево на 2 сигма и вправо на 2, значит, процент значений в пределах $158см \leq x \leq 190 см$ равен $95.4\%$

**е). не выше 150 см или не ниже 190 см**  
$ Z = \frac{150-174}{8}= -3 $
Смотрим по таблице значение  по [-3.0] и [0.00], $P(x \leq 150)= 0.0014$  
Добавим сюда вероятность из пункта **(б)**  
$P(150 см \leq x \leq 190 см)=0.0014 + 0.0228=0.0242$  
По правилу 3-х сигм - это отрезок $ (-\infty ; -3\sigma] \cup[+2\sigma;+ \infty)$  
$(100\%- 99.72\%) $(вне 3х $\sigma$) $+ (\frac{99.72\%-95.4\%}{2})$(добавляем $[+2\sigma;+3\sigma]$)$ = 0.28\%+2.16\%=2.44\%$ .


**ё). не выше 150 см или не ниже 198 см**  
$ Z = \frac{198-174}{8}= 3 $  
Снова попадаем в отрезок, кратный  $\sigma$, это $ (-\infty ; -3\sigma] \cup[+3\sigma;+ \infty)$ ,а такое мы уже расписали в пункте **(е)**:  
$(100\%- 99.72\%) = 0.28\% $


**ж). ниже 166 см.**  
Берем найденную по таблице вероятность в пункте **(в)** : $P(x \leq 166)= 0.1587$
Или по 3-м сигмам:  
50%(слева от $\mu$) $- \frac{68\%}{2} = 16\%$

### Задача 5

*На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой $M(X) = 178$ см и $D(X) = 25$ кв.см?*

Если считать, что рост - это величина с нормальным распределением, то можем использовать ту же формулу оценки отклонения:  
$$ Z = \frac{X-\mu}{\sigma} $$ 
где,  по условию задачи :  
$\sigma = \sqrt{D(X)} = \sqrt{25}= 5$ см,  
$\mu= M(X) = 178$ см  
$X=190$ см$

$$ Z = \frac{190-178}{5}= 2.4 $$ 

Ответ: отклонение в 2.4 сигмы, и шанс найти человека такого роста в этой популяции меньше 0,82%
