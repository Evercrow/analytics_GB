```python
a = np.array([12, 10, 11, 19, 13, 11, 17, 15, 19, 14, 21, 18, 21, 11, 17, 14, 15, 17, 20, 19])
b = np.array([11, 13, 18, 15, 17, 18, 10, 21, 26, 15, 11, 12, 15, 17, 10, 18, 18, 12,21, 20])
sts.ttest_ind(a,b)
```

```python
sts.t.ppf(0.8737549039369696/2,2*(len(a)-1))
```

Используя функцию stats.ttest_ind, проведите односторонний тест. Проверить ,что мю 1 > мю 2


```python
sts.ttest_ind(a,b, alternative = "greater")
```

Так как pvalue=0.56 > alpha=0.05 , то верна нулевая гипотеза , то есть мю1 = мю2 

Используя функцию stats.ttest_ind, проведите односторонний тест. Проверить ,что мю 2 > мю 1


```python
sts.ttest_ind(b,a, alternative = "less")
```

Так как pvalue=0.56 > alpha=0.05 , то верна нулевая гипотеза , то есть мю1 = мю2
